```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
cancer.keys()
import pandas as pd
df = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
df
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
scaled_df = std_scaler.fit_transform(df)
scaled_df
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=4)
x = pca.fit_transform(scaled_df)
```

```python
x
```

```python
principal_breast_Df = pd.DataFrame(data =x
             , columns = ['principal component 1', 'principal component 2','principal component 3','principal component 4'])
```

```python
principal_breast_Df
```

```python
print('Explained variation per principal component: {}'.format(pca.explained_variance_ratio_))
```

```python
from sklearn.model_selection import train_test_split
y = cancer["target"]

X_train, X_test, y_train, y_test = train_test_split(principal_breast_Df, y, test_size=0.3,random_state=42)
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression( )
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```

```python
from sklearn.metrics import confusion_matrix
print( confusion_matrix(y_test, y_pred))
     
```

```python
from sklearn import  metrics
print(
    f"Classification report for classifier {clf}:\n"
    f"{metrics.classification_report(y_test, y_pred)}\n"
)
```
